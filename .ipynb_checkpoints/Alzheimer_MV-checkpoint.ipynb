# 1. Uvod

Alzheimerova bolest je progresivna bolest koja pogađa neurone u ljudskom mozgu. Posljedice napretka ove bolesti su gubitak pamćenja, paranoja, nesnalaženje u prostoru i vremenu, manjak koncentracije itd. Bolest većinom zahvaća starije stanovništvo (65 i više godina starosti). 

Rano otkrivanje ove bolesti dovodi do pronalaska odgovarajuće terapije za pacijenta i sprječavanja ili usporavanja nastanka velikog broja oštećenja neurona.

Ovaj projekt bavi se analizom podataka koji bi mogli pomoći u ranom otkrivanju bolesti. Zaključci i predviđanja doneseni su na temelju javnog skupa [podataka](https://www.kaggle.com/datasets/jboysen/mri-and-alzheimers) koji je omogućio projekt Open Access Series of Imaging Studies (OASIS).

# 2. Deskriptivna statistika

Ovaj projekt koristi longitudinalne MR podatke. Podaci su skupljeni od 150 induvidualnih sudionika dobi između 60 i 69 godina. Svaki sudionik bio je skeniran dva ili više puta s razmakom od minimalno godine dana. Time su se dobila 373 različita $T_{1}$ MR skena.

Opis značajki:
- Subject ID = identifikacija pojedinog subjekta
- MRI ID = identifikacija pojedinog MRI pregleda
- Group = klasa
    - Converted = inicijalno smatrani nedementnim, kasnije klasificirani kao dementni
    - Demented = dementni
    - Nondemented = nedementni
- Visit = redni broj skena za pojedini subjekt
- MR Delay = MR vrijeme odgode
- M/F = spol ispitanika
    - M = muškarci
    - F = žene
- Hand = dominantna ruka subjekta
- Age = dob subjekta
- EDUC = broj godina edukacije
- SES = socioekonomski status (1 je najviši, a 5 najniži status)
- MMSE = mali mentalni test (0 je najgori, a 30 najbolji rezultat)
- CDR = procjena demencije
    - 0 = nedementan
    - 0.5 = jako blagi AD*
    - 1 = blagi AD
    - 2 = umjeren AD
    - 3 = jak AD
- eTIV = procjena ukupnog volumena lubanje (lubanja, mozak, moždane ovojnice, cerebrospinalni likvor) u mm3
- nWBV = normaliziran volumen cijelog mozga
- ASF = Atlas Scaling Factor, faktor za usporedbu procijenjenog volumena lubanje, eTIV-a

AD = Alzheimer disease = Alzheimerova bolest

### 2.4.4. Podjela i normalizacija podataka prije provođenja algoritama strojnog učenja

Predikcija stanja ispitanika provodit će se na temelju dostupnih značajki. Prije implementacije modela i treniranja, značajke se moraju normalizirat kako bi se spriječilo krivo zaključivanje utjecajnosti pojedine značajke. Značajke se translatiraju i smještaju na istu skalu između 0 i 1.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

y = data['Group'].values #zavisna varijabla
X = data[['M/F', 'Age', 'EDUC', 'SES', 'MMSE', 'eTIV', 'nWBV', 'CDR']] #nezavisne varijable

#podjela skupa podataka na podatke za treniranje i testiranje u omjeru 75:25
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

#normalizacija s obzirom na podatke za treniranje
scaler = MinMaxScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# 3. Modeli predikcije

U ovom istraživanju isprobano je nekoliko klasifikacijskih modela, točnije Decision Tree, Random Forest, SVM, XGBoost i Voting. Performance svakog modela i njihove usporedbe bit će ispitane i prikazane točnošću, preciznošću, recallom, F1 scoreom i matricom zabune.

Modeli su zapravo skup hipoteza definiranih do na parametre. Učenje modela je pronalaženje parametara/ hipoteze koja je optimalna, tj. daje najbolje predviđanje na neviđenim (testnim) podacima.
Modeli imaju i svoje hiperparametre (npr. stupanj polinoma) koje je također potrebno odrediti. Ovaj dio ne odrađuju algoritmi strojnog učenja, već je to posao istraživača. Kako bi se odabrali optimalni hiperparametri izvodi se postupak cross-validacije.

Cross-validacija je postupak u kojem se skup za treniranje dijeli na k dijelova. k-1 od tih se koristi za treniranje modela, a samo jedan se koristi za validaciju i to se ponavlja k puta. Validacija zapravo simulira proces testiranja te pomaže odabrati optimalne hiperparametre.
Nakon odabranih hiperparametra, moguće je konačno provesti testiranje modela.

## 3.1. Pregled mjera kvalitete modela

### 3.1.1. Točnost (accuracy)

Točnost mjeri koliko je model dobro označio podatke u odnosu na stvarni svijet, tj. mjeri omjer točno klasificiranih primjera i svih ostalih.

$$
  accuracy (in percentage) = \frac{TP + TN}{TP + FP + TN + FN}
$$

### 3.1.2 Preciznost (precision)

Preciznost mjeri izvedbu pojedine klase, tj. npr. omjer točno predviđenih pozitivnih primjera i ukupno pozitivno predviđenih primjera.

$$
  precision (in percentage) = \frac{TP}{TP + FP}
$$

Ako je preciznost 1, znači da je klasifikator dobar.

### 3.1.3. Odaziv (recall)

Recall je omjer uzoraka klase koja je točno predviđena.
$$
  recall (in percentage) = \frac{TP}{TP + FN}
$$

### 3.1.4. F1 Score 

F1 Score predstavlja harmonijsku sredinu preciznosti i recalla.

$$
  F1Score (in percentage) = \frac{2*Precision*Recall}{Precision + Recall}
$$

### 3.1.5 Matrica zabune 

Matrica zabune je tablična vizualizacija dobivenih predikcija u odnosu na anotirane oznake. Dijagonalni elementi matrice su točne predikcije za različite klase, a elementi van dijagonala predstavljaju broj uzoraka koji su krivo klasificirani.

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

## 3.2. Decision Tree model

Decision Tree je jednostavna metoda strojnog učenja koja uči jednostavna pravila odlučivanja na temelju podataka za treniranje. Stablo se dijeli s obzirom na granične vrijednosti, npr. entropiju. Složenost stabla određuje njegova dubina. Što je stablo dublje, to je model složeniji. 

Prednosti stabla oluke je njegova jednostavna interpretacija i vizualizacija, a nedostatak je lako postizanje prenaučenosti i nestabilnosti.

Hiperparametar stabla odluke je njegova dubina.

```python
param_dt= {
    "max_depth":range(1,9),
}

#optimizacija hiperparametra
model_dt= DecisionTreeClassifier(criterion='entropy', random_state=42)
dt_grid = GridSearchCV(estimator = model_dt, param_grid = param_dt, cv = 5, 
                        n_jobs = 4, scoring='f1')

dt_grid.fit(X_train_scaled, y_train)
print(dt_grid.best_estimator_)

#odabir najboljeg modela
model_dt = dt_grid.best_estimator_
model_dt.fit(X_train_scaled,y_train)

#iscrtavanje stabla odluke
plt.figure(figsize=(8, 5))
tree.plot_tree(model_dt, feature_names=data.columns, class_names=['Non-demented','Demented'], filled=True)
plt.title("Dobiveno stablo odluke")

dtm_predicted_y = model_dt.predict(X_test_scaled)

#mjere točnosti
dtm_test_accuracy = model_dt.score(X_test_scaled, y_test)
dtm_test_precision = precision_score(y_test, dtm_predicted_y, pos_label=1, average='binary')
dtm_test_recall = recall_score(y_test, dtm_predicted_y, pos_label=1, average='binary')
dtm_test_f1score = f1_score(y_test, dtm_predicted_y, pos_label=1, average='binary')
dtm_confusion_matrix = confusion_matrix(y_test, dtm_predicted_y)

print("Test accuracy: ", dtm_test_accuracy)
print("Test precision: ", dtm_test_precision)
print("Test recall:  ", dtm_test_recall)
print("Test F1Score:  ", dtm_test_f1score, "\n")

dtm_cm_display = ConfusionMatrixDisplay(confusion_matrix = dtm_confusion_matrix, display_labels = [False, True])
dtm_cm_display.plot()
for labels in dtm_cm_display.text_.ravel():
    labels.set_fontsize(30)
plt.title("Matrica zabune")
```

## 3.3.  Random forest classifier

Random forest classifier je skup stabala odluke od kojih svaki dobiva podskup podataka, a izlaz šume je rezultat većinskog glasanja. Ovime se nastoji spriječiti prenaučenosti koja je moguća dubokim stablom odluke.

Bitni hiperparametri su n_estimators, tj. broj stabala u šumi, max_depth kao maksimalna dubina stabala i max_features koji predstavlja broj značajki.

```python
param_rfc= {
    "n_estimators": range(1,25),
    "max_features": range(1,9),
    "max_depth": range(1,9)
}

#optimizacija hiperparametra
model_rfc= RandomForestClassifier(criterion='entropy', random_state=42)
rfc_grid = GridSearchCV(estimator = model_rfc, param_grid = param_rfc, cv = 5, 
                            n_jobs = 4, scoring='f1')

rfc_grid.fit(X_train_scaled, y_train)
print(rfc_grid.best_estimator_)

#odabir najboljeg modela
model_rfc = dt_grid.best_estimator_
model_rfc.fit(X_train_scaled,y_train)

rfc_predicted_y = model_rfc.predict(X_test_scaled)

#mjere točnosti
rfc_test_accuracy = model_rfc.score(X_test_scaled, y_test)
rfc_test_precision = precision_score(y_test, rfc_predicted_y, pos_label=1, average='binary')
rfc_test_recall = recall_score(y_test,rfc_predicted_y, pos_label=1, average='binary')
rfc_test_f1score = f1_score(y_test, rfc_predicted_y, pos_label=1, average='binary')
rfc_confusion_matrix = confusion_matrix(y_test, rfc_predicted_y)

print("Test accuracy: ", rfc_test_accuracy)
print("Test precision: ", rfc_test_precision)
print("Test recall:  ", rfc_test_recall)
print("Test F1Score:  ", rfc_test_f1score)

rfc_cm_display = ConfusionMatrixDisplay(confusion_matrix = rfc_confusion_matrix, display_labels = [False, True])
rfc_cm_display.plot()
for labels in rfc_cm_display.text_.ravel():
    labels.set_fontsize(30)
plt.title("Matrica zabune")
```

## 3.4. SVM

SVM je metoda koja pokušava naći granicu između dvije klase u prostoru takvu da ta granica bude maksimalno udaljena od najbližih joj primjera. SVM pretpostavlja linearnu odvojivost klasa, no moguće je dodati i hiperparametar C koji utječe na složenost modela te omogućuje klasifikaciju i linearno neodvojivih primjera.

Osim C-a, moguće je odabrati i jezgru modela pomoću koje se računa udaljenost između primjera te gammu koja se koristi za rbf jezgru za određivanje preciznosti.

```python
param_svm= {
    "C": [0.001, 0.01, 0.1, 0.15, 0.5, 1, 5, 10, 100, 1000, 1500],
    "gamma":[0.001, 0.01, 0.1, 1, 1.5, 2, 10, 100, 1000],
    "kernel": ['rbf', 'linear', 'poly', 'sigmoid']
}

#optimizacija hiperparametra
model_svm= SVC()
svm_grid = GridSearchCV(estimator = model_svm, param_grid = param_svm, cv = 5, 
                               verbose=0,n_jobs = 4, scoring='f1')

svm_grid.fit(X_train_scaled, y_train)
print(svm_grid.best_estimator_)

#odabir najboljeg modela
model_svm = svm_grid.best_estimator_
model_svm.fit(X_train_scaled,y_train)
model_svm.score(X_test_scaled,y_test)

svm_predicted_y = model_svm.predict(X_test_scaled)

#mjere točnosti
svm_test_accuracy = model_svm.score(X_test_scaled, y_test)
svm_test_precision = precision_score(y_test, svm_predicted_y, pos_label=1, average='binary')
svm_test_recall = recall_score(y_test, svm_predicted_y, pos_label=1, average='binary')
svm_test_f1score = f1_score(y_test, svm_predicted_y, pos_label=1, average='binary')
svm_confusion_matrix = confusion_matrix(y_test, svm_predicted_y)

print("Test accuracy: ", svm_test_accuracy)
print("Test precision: ", svm_test_precision)
print("Test recall:  ", svm_test_recall)
print("Test F1Score:  ", svm_test_f1score)

svm_cm_display = ConfusionMatrixDisplay(confusion_matrix = svm_confusion_matrix, display_labels = [False, True])
svm_cm_display.plot()
for labels in svm_cm_display.text_.ravel():
    labels.set_fontsize(30)
plt.title("Matrica zabune")
```

## 3.5. XGBoost

XGBoost je model strojnog učenja koji implementira nekoliko stabala odluke, a fokusira se na performanse. Za razliku od modela nasumične šume, koji paralelno obrađuje stabla, XGBoost izgrađuje ih slijedno i svako stablo nastoji ispraviti pogreške prethodnog.

Hiperparametri koji se moraju odrediti su dubina stabala, stopa učenja i broj parametara koji određuje broj iteracija.

```python
param_xgb = {
    "learning_rate": [0.0000005, 0.0001, 0.001, 0.01, 0.025, 0.005,0.5, 0.075, 0.1, 0.15, 0.2, 1],
    "max_depth":range(1,9),
    "n_estimators":range(1,25)
    }

#optimizacija hiperparametra
model_xgb= XGBClassifier(eval_metric='logloss')
xgb_grid = GridSearchCV(estimator = model_xgb, param_grid = param_xgb, cv = 5, 
                            n_jobs = 4, scoring='f1')

xgb_grid.fit(X_train_scaled, y_train)
print(xgb_grid.best_estimator_)

#odabir najboljeg modela
model_xgb = xgb_grid.best_estimator_
model_xgb.fit(X_train_scaled,y_train)
model_xgb.score(X_test_scaled,y_test)

xgb_predicted_y = model_xgb.predict(X_test_scaled)

#mjere točnosti
xgb_test_accuracy = model_xgb.score(X_test_scaled, y_test)
xgb_test_precision = precision_score(y_test, xgb_predicted_y, pos_label=1, average='binary')
xgb_test_recall = recall_score(y_test, xgb_predicted_y, pos_label=1, average='binary')
xgb_test_f1score = f1_score(y_test, xgb_predicted_y, pos_label=1, average='binary')
xgb_confusion_matrix = confusion_matrix(y_test, xgb_predicted_y)

print("Test accuracy: ", xgb_test_accuracy)
print("Test precision: ", xgb_test_precision)
print("Test recall:  ", xgb_test_recall)
print("Test F1Score:  ", xgb_test_f1score)

xgb_cm_display = ConfusionMatrixDisplay(confusion_matrix = xgb_confusion_matrix, display_labels = [False, True])
xgb_cm_display.plot()
for labels in xgb_cm_display.text_.ravel():
    labels.set_fontsize(30)
plt.title("Matrica zabune")
```

## 3.6. Voting

Ideja Voting modela je kombiniranje drugačijih modela strojnog učenja i korištenje njihovih izlaza za odluku o klasi. Kod hard votinga, klasa primjera za testiranje određuje se prema najčešćoj predviđenoj klasi različitih modela. S druge strane, kod soft votinga ona se određuje pronalaskom argmax-a vjerojatnosti klasa.

U ovom istraživanju korišteni su klasifikatori logistička regresija, KNN i naivni Bayesov klasifikator.

Hard Voting:

```python
#klasifikatori
clf1 = LogisticRegression(max_iter=1000)
clf2 = KNeighborsClassifier()
clf3 = GaussianNB()

param_hardv = {
    'lr__C': [0.01, 0.1,0.25, 0.5, 1, 10],
    'knn__n_neighbors': range(2,30),
    'gnb__var_smoothing': [0.01, 1e-1,1e-09, 1e-08, 1e-07],
}

#optimizacija hiperparametra
voting_model_h = VotingClassifier(estimators=[ ('lr', clf1), ('knn', clf2), ('gnb', clf3)], voting='hard')
hv_grid = GridSearchCV(voting_model_h, param_hardv, n_jobs=4, cv=5, scoring='f1')

hv_grid.fit(X_train_scaled, y_train)
print(hv_grid.best_estimator_)

#odabir najboljeg modela
voting_model_hv = hv_grid.best_estimator_
voting_model_hv.fit(X_train_scaled, y_train)

vot_predicted_y = voting_model_hv.predict(X_test_scaled)

v_test_accuracy_h = voting_model_hv.score(X_test_scaled, y_test)
v_test_precision_h = precision_score(y_test,  vot_predicted_y, pos_label=1, average='binary')
v_test_recall_h = recall_score(y_test, vot_predicted_y, pos_label=1, average='binary')
v_test_f1score_h = f1_score(y_test, vot_predicted_y, pos_label=1, average='binary')
v_confusion_matrix_h = confusion_matrix(y_test, vot_predicted_y)

print("Test accuracy: ", v_test_accuracy_h)
print("Test precision: ", v_test_precision_h)
print("Test recall:  ", v_test_recall_h)
print("Test F1Score:  ", v_test_f1score_h)

vh_cm_display = ConfusionMatrixDisplay(confusion_matrix = v_confusion_matrix_h, display_labels = [False, True])
vh_cm_display.plot()
for labels in vh_cm_display.text_.ravel():
    labels.set_fontsize(30)
plt.title("Matrica zabune")
```

Soft Voting:

```python
#klasifikatori
clf1 = LogisticRegression(max_iter=1000)
clf2 = KNeighborsClassifier()
clf3 = GaussianNB()

param_soft = {
    'lr__C': [0.01, 0.1,0.25, 0.5, 1, 10],
    'knn__n_neighbors': range(2,30),
    'gnb__var_smoothing': [0.01, 0.1,1e-09, 1e-08, 1e-07],
}

#optimizacija hiperparametra
voting_model_s = VotingClassifier(estimators=[('lr', clf1), ('knn', clf2), ('gnb', clf3)], voting='soft')
sv_grid = GridSearchCV(voting_model_s, param_soft, n_jobs=4, cv=5, scoring='f1')

sv_grid.fit(X_train_scaled, y_train)
print(sv_grid.best_estimator_)

#odabir najboljeg modela
voting_model_sv = sv_grid.best_estimator_
voting_model_sv.fit(X_train_scaled, y_train)

vot_predicted_y = voting_model_sv.predict(X_test_scaled)

v_test_accuracy_s = voting_model_sv.score(X_test_scaled, y_test)
v_test_precision_s = precision_score(y_test,  vot_predicted_y, pos_label=1, average='binary')
v_test_recall_s = recall_score(y_test, vot_predicted_y, pos_label=1, average='binary')
v_test_f1score_s = f1_score(y_test, vot_predicted_y, pos_label=1, average='binary')
v_confusion_matrix_s = confusion_matrix(y_test, vot_predicted_y)

print("Test accuracy: ", v_test_accuracy_s)
print("Test precision: ", v_test_precision_s)
print("Test recall:  ", v_test_recall_s)
print("Test F1Score:  ", v_test_f1score_s)

vs_cm_display = ConfusionMatrixDisplay(confusion_matrix = v_confusion_matrix_s, display_labels = [False, True])
vs_cm_display.plot()
for labels in vs_cm_display.text_.ravel():
    labels.set_fontsize(30)
plt.title("Matrica zabune")
```
