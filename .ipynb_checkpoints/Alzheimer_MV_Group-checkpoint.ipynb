# 1. Uvod

Alzheimerova bolest je progresivna bolest koja pogađa neurone u ljudskom mozgu. Posljedice napretka ove bolesti su gubitak pamćenja, paranoja, nesnalaženje u prostoru i vremenu, manjak koncentracije itd. Bolest većinom zahvaća starije stanovništvo (65 i više godina starosti). 

Rano otkrivanje ove bolesti dovodi do pronalaska odgovarajuće terapije za pacjenta i sprječavanja ili usporavanja nastanka velikog broja oštećenja neurona.

Ovaj projekt bavi se analizom podataka koji bi mogli pomoći u ranom otkrivanju bolesti. Zaključci i predviđanja doneseni su na temelju javnog skupa podataka koji je omogućio projekt Open Access Series of Imaging Studies (OASIS).

# 2. Deskriptivna statistika

Ovaj projekt koristi longitudinalne MR podatke. Podaci su skupljeni od 150 induvidualnih sudionika dobi između 60 i 69 godina. Svatki sudionik bio je skeniran dva ili više puta s razmakom od minimalno godine dana. Time su se dobila 373 različita $T_{1}$ MR skena. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

import warnings
warnings.filterwarnings('ignore')
```

Učitavanje dostupnih podataka:

```python
data = pd.read_csv("oasis_longitudinal.csv")
print("Podaci imaju {} redova i {} stupaca".format(data.shape[0], data.shape[1]))
```

```python
# prikaz skupa podataka
data.head()
```

```python
print("Podaci imaju sljedeće značajke: {}".format(data.columns.values))
```

Opis značajki:
- Subject ID = identifikacija pojedinog subjekta
- MRI ID = identifikacija pojedinog MRI pregleda
- Group = klasa (Converted/Demented/Nondemented)
- Visit = redni broj skena za pojedini subjekt
- MR Deley = MR vrijeme odgode
- M/F = spol ispitanika
- Hand = dominantna ruka subjekta
- Age = dob subjekta
- EDUC = broj godina edukacije
- SES = socioekonomski status (1 je najviši, a 5 najniži status)
- MMSE = mali mentalni test (0 je najgori, a 30 najbolji rezultat)
- CDR = procjena demencije (0 = nedementan, 0.5 = jako blagi AD, 1 = blagi AD, 2 = umjeren AD, 3 = jak AD)
- eTIV = procjena ukupnog volumena lubanje (lubanja, mozak, moždane ovojnice, cerebrospinalni likvor) u mm3
- nWBV = normaliziran volumen cijelog mozga
- ASF = faktor za usporedbu procjenjenog volumena lubanje, eTIV-a (Atlas Scaling Factor)

Prije analize podataka, izbačeni su svi podaci koji se ne odnose na prve skenove jer je cilj projekta odrediti predikciju rane faze Altzheimerove bolesti, tj. donijeti zaključak već nakon prvog pregleda, tj. skena.

```python
data = data.loc[data.Visit==1]  #ostavljanje samo prvih skenova
data = data.reset_index(drop=True)
print("Podaci imaju {} redova i {} stupaca".format(data.shape[0], data.shape[1]))
```

### 2.4.2 Odnos značajki i ciljane značajke (Group)

#### Odnos godina i spola  na stanje demencije:

```python
fig, ax = plt.subplots(1,2, figsize=(20,10))
sns.boxplot(ax = ax[0], x="Group", y="Age", data=data)
ax[0].set_title("Odnos stanja i godina")
sns.violinplot(ax = ax[1], x="Group", y="Age", hue="M/F", data=data, palette="muted", split=True, inner="quartile")
ax[1].set_title("Odnos stanja i godina po spolu")
```

Vidljivo je da je interval godina nedemetnih ispitanika veći od onog dementnih, ali i da je u intervalu od 70-80 godina sadržana većina dementnih subjekata. Ovo je objašnjivo smrtnošću dementnih osoba koje nemaju priliku doživjeti duboku starost. Usporedbom po spolovima, vidljivo je da su žene u ranijoj dobi dijagnosticirane demencijom u donosu na muškarce.

```python
mask_demented = data.Group == 1

count_per_gender_demented = data.loc[mask_demented,:].groupby('M/F').count().loc[:,'Group'].values
count_per_gender_nondemented = data.loc[~mask_demented,:].groupby('M/F').count().loc[:,'Group'].values
gender = data.groupby('M/F').count().index.values

# plot
plt.figure(figsize = (7,5))
plt.bar(gender, count_per_gender_nondemented, color='green', label='nondemented')
plt.bar(gender, count_per_gender_demented, color='red', label='demented', bottom=count_per_gender_nondemented)
plt.xlabel("Spol")
plt.ylabel("Količina")
plt.legend(loc='best')
plt.grid()
plt.show()
```

U istraživanju je sudjelovalo više žena, no većina ih je bila nedementna. S druge strane, više od polovice muškaraca bilo je proglašeno dementnim.

#### Odnos godina obrazovanja na stanje demencije:

```python
plt.figure(figsize = (7,5))
sns.boxplot(x="Group", y="EDUC", data=data)
plt.title("Odnos godina obrazovnja i stanja")
```

Ispitainici proglašeni nedementnim su se više godina obrzovali (u prosjeku oko 16 godina), dok dementni imaju iza sebe manji broj godina obrazovanja.

#### Odnos socioekonomskog statusa na stanje demencije:

```python
count_per_ses_demented = data.loc[mask_demented,:].groupby('SES').count().loc[:,'Group'].values
count_per_ses_nondemented = data.loc[~mask_demented,:].groupby('SES').count().loc[:,'Group'].values
ses = data.groupby('SES').count().index.values

# plot
plt.figure(figsize = (7,5))
plt.bar(ses, count_per_ses_nondemented, color='green', label='nedementni')
plt.bar(ses, count_per_ses_demented, color='red', label='dementni', bottom=count_per_ses_nondemented)
plt.xlabel("Socioekonomski status")
plt.ylabel("Količina")
plt.title("Raspodjela demencije po socioekonomskom statusu")
plt.legend(loc='best')
plt.grid()
plt.show()
```

1 je najviši status i tu ima više dementnih, ali to je očekivano jer bogatiji i žive duže pa su ispitanici stariji (vidi poglavlje 2.4.3), dok recimo statusa 2 ima više nedementnih koji si mogu priuštiti dobar zdravstveni život. ### 2.4.4. Podjela i normalizacija podataka prije provođenja algoritama strojnog učenja

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

y = data['Group'].values #zavisna varijabla
X = data[['M/F', 'Age', 'EDUC', 'SES', 'MMSE', 'eTIV', 'nWBV']] #nezavisne varijable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)

scaler = MinMaxScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Napomena: u predloženomo radu, autori nisi izbacili značajke ASF i CDR.
